```python
Rp = []
Rp.append(res.Resistance("$R'_{cond}$","W/m"))
Rp.append(res.Resistance("$R'_{cont}$","W/m"))
Rp.append(res.Resistance("$R'_{conv}$","W/m"))
```

```python
ro = 75e-3
ri = 25e-3
Ti = 5.
Tinf = -10
To = 25.
k = 10.
h = 100
Rtc = 0.01

Rp[0].cond_cylinder(k,ri,ro)
Rp[1].contact(0.01,1)
Rp[2].convection(h,2*np.pi*ro)
q_conv = (To - Tinf)/Rp[2].R
q_cond_wo_contact = (To - Ti)/Rp[0].R
q_cond_w_contact = (To - Ti)/(Rp[0].R + Rp[1].R)
print(q_conv + q_cond_wo_contact)
print(q_conv + q_cond_w_contact)
```
