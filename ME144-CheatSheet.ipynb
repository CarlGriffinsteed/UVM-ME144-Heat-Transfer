## A word about radiation

When dealing with radiation between a surface at temperature $T_s$ and the atmosphere, the far away surface temperature is often defined as $T_{sky}$ instead of $T_{sur}$:

$$
q_{rad}= h_rA(T_s-T_{sky})
$$

with

$$
h_r = \varepsilon\sigma(T_s^2+T_{sky}^2)(T_s+T_{sky})
$$

## Equations

### Convection
#### Heat flux
$$
q''_{i} = \frac{1}{R''_{conv,x}}(T_{s,x}-T_{\infty,x})
$$

$$
R''_{conv,x}=\frac{1}{h_x}
$$
#### Heat rate per unit length
$$
q'_{i} = \frac{1}{R'_{conv,x}}(T_{s,x}-T_{\infty,x})
$$

$$
R'_{conv,x}=\frac{1}{h_xL}
$$
#### Heat rate
$$
q_{i} = \frac{1}{R_{conv,x}}(T_{s,x}-T_{\infty,x})
$$

$$
R_{conv,x}=\frac{1}{h_xA}
$$

### Conduction heat transfer:
#### Plane wall (heat flux)
$t$: wall thickness
$$
q''_{cond}= \frac{1}{R''_{cond}}(T_{a}-T_{b})
$$

$$
R''_{cond}=\frac{t}{k}
$$
#### Plane wall (heat rate per unit length)
$L$: height $H$ or width $W$
$$
q'_{cond}= \frac{1}{R''_{cond}}(T_{a}-T_{b})
$$

$$
R'_{cond}=\frac{t}{kL}
$$
#### Plane wall (heat rate)
$A=H\times W$


$$
q_{cond}= \frac{1}{R_{cond}}(T_{a}-T_{b})
$$

$$
R_{cond}=\frac{t}{kA}
$$
#### Cylindrical shell (heat rate per unit length)
$r_i,r_o$: inner and outer radii
$$
R'_{cond}=\frac{\ln(r_o/r_i)}{2\pi k}
$$
#### Cylindrical shell (heat rate)
$L$: shell length
$$
R_{cond}=\frac{\ln(r_o/r_i)}{2\pi k L}
$$
#### Spherical shell
$$
R_{cond}=\frac{1}{4\pi k}\left(\frac{1}{r_i}-\frac{1}{r_o}\right)
$$

### Conservation of energy
Time variation of **stored** energy = **in**flux of thermal energy - **out**flux of thermal energy + **generated** energy
$$
\frac{dE_{st}}{dt}=\dot{E}_{in}-\dot{E}_{out}+\dot{E}_{g}
$$
$V$:volume
$$
\rho C_p V \frac{dT}{dt}=q_{in}-q_{out}+q_g
$$
